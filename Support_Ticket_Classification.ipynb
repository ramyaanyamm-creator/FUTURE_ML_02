```python
# ============================================
# FULL SETUP CODE
# ============================================

# Import Libraries
import pandas as pd
import numpy as np
import re
import nltk
import zipfile
import os
import pickle

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)

# ============================================
# DOWNLOAD NLTK DATA
# ============================================

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# ============================================
# EXTRACT ZIP FILE
# ============================================

with zipfile.ZipFile("archive (1).zip", 'r') as zip_ref:
    zip_ref.extractall("dataset_folder")

print("ZIP File Extracted Successfully")

# ============================================
# CHECK FILES
# ============================================

print(os.listdir("dataset_folder"))

# ============================================
# LOAD DATASET
# ============================================

df = pd.read_csv("dataset_folder/customer_support_tickets.csv")

print(df.head())

# ============================================
# REMOVE MISSING VALUES
# ============================================

df = df.dropna()

# ============================================
# TEXT CLEANING
# ============================================

stop_words = set(stopwords.words('english'))

def clean_text(text):

    # Lowercase
    text = text.lower()

    # Remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenization
    words = word_tokenize(text)

    # Remove stopwords
    words = [word for word in words if word not in stop_words]

    return " ".join(words)

# Apply Cleaning
df['cleaned_text'] = df['Ticket Description'].apply(clean_text)

print(df[['Ticket Description', 'cleaned_text']].head())

# ============================================
# TF-IDF FEATURE EXTRACTION
# ============================================

tfidf = TfidfVectorizer(max_features=5000)

X = tfidf.fit_transform(df['cleaned_text'])

print(X.shape)
```

```python
y_category = df['Ticket Type']
print(y_category.head())
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_category,
    test_size=0.2,
    random_state=42
)
print("Training Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)
from sklearn.linear_model import LogisticRegression
category_model = LogisticRegression(max_iter=1000)
category_model.fit(X_train, y_train)
print("Category Model Trained Successfully")
y_pred = category_model.predict(X_test)
print("\nCategory Predictions:")
print(y_pred[:10])

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)
accuracy = accuracy_score(y_test, y_pred)
print("\n===================================")
print("CATEGORY PREDICTION RESULTS")
print("===================================")
print("\nCategory Prediction Accuracy:")
print(accuracy)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
y_priority = df['Ticket Priority']
print("\nPriority Values:")
print(y_priority.head())
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(
    X,
    y_priority,
    test_size=0.2,
    random_state=42
)
print("\nPriority Training Shape:", X_train_p.shape)
print("Priority Testing Shape:", X_test_p.shape)
priority_model = LogisticRegression(max_iter=1000)
priority_model.fit(X_train_p, y_train_p)
print("\nPriority Model Trained Successfully")
y_pred_p = priority_model.predict(X_test_p)
print("\nPriority Predictions:")
print(y_pred_p[:10])
accuracy_p = accuracy_score(y_test_p, y_pred_p)
print("PRIORITY PREDICTION RESULTS")
print("\nPriority Prediction Accuracy:")
print(accuracy_p)
print("\nClassification Report:")
print(classification_report(y_test_p, y_pred_p))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test_p, y_pred_p))
custom_ticket = [
    "My payment failed and money was deducted"
]
cleaned_custom = [clean_text(text) for text in custom_ticket]
custom_vector = tfidf.transform(cleaned_custom)
predicted_category = category_model.predict(custom_vector)
predicted_priority = priority_model.predict(custom_vector)
print("\nTicket:")
print(custom_ticket[0])
print("\nPredicted Category:")
print(predicted_category[0])
print("\nPredicted Priority:")
print(predicted_priority[0])
import pickle
with open("category_model.pkl", "wb") as file:
    pickle.dump(category_model, file)
with open("priority_model.pkl", "wb") as file:
    pickle.dump(priority_model, file)
with open("tfidf_vectorizer.pkl", "wb") as file:
    pickle.dump(tfidf, file)
```
